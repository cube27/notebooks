2. What is Quantopian?

  - Platform that allows users to **develop**, **test**, and **deploy**
    quantitative financial algorithms written in Python.

  - Community where quants can share ideas, get help, and collaborate on
    algorithms.

  - What Quantopian currently good at?

    - Allowing people to write and test strategies that they already have.

    - Allowing people to connect those strategies to the market to trade with
      real money.

  - What are we less good at?

    - Providing tools for figuring out what my strategy should be in the first
      place.

      - Interactively exploring/visualizing relationships in data.
      - Computations that take more than a minute.

    - Allowing users to share stuff that isn't Python code or backtest results.

    - Supporting ad-hoc incorporation of different data sources.
      - `fetch_csv()` gets you some of the way there, but again, not great for
        figuring out what your inputs should be in the first place.

5. IPython Notebook:

   - What we just saw is awesome as an interactive shell, but not the best for:

     1. Saving/Reusing Work:
       - Don't want to have to keep a session open forever.
       - No good way to share a session.
       - `%edit` helps for writing larger chunks of code.

     2. Minimal support for rich display.

     3. Sometimes you want to include other content besides Python code
     (e.g. instructions, annotations, diagrams...).

   - Quit IPython, run `ipython notebook --matplotlib=inline`.

     - This runs a **notebook server**, which I can connect to by pointing my
       browser (or another application that knows how to read the format) to

8. Research Env Sneak Peek:

   - We've seen that IPython Notebook acts as a nice middle ground between the
     static nature of a script/library and the dynamic (but impermanent) nature
     of a shell session.

   - We've also seen that IPython Notebook enables sharing computation, which
     includes not just Python code but also data visualizations, documentation,
     and other content.

   - And we've seen that we can use IPython Notebook with the state of the art
     tools (like `pandas` and `zipline`) to get even more out of these tools.
